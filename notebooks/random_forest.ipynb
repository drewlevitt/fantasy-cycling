```python
df = pd.get_dummies(df.drop(['Rider ID', 'Rider Name'], axis = 1)).fillna(0)
train = df[df['Observation Year'] < 2019].drop('Observation Year', axis = 1)
test = df[df['Observation Year'] == 2019].drop('Observation Year', axis = 1)

X = train.drop('Score 0ya', axis = 1).values
y = train['Score 0ya'].values
```
